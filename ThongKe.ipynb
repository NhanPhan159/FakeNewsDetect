```python
fake = pd.read_csv('fake_news_viet.csv')
real = pd.read_csv('real_news_viet.csv')

total = pd.concat([fake,real],ignore_index=True)
loai = total['nhan'].value_counts().values
plt.bar(["thật","giả"],[loai[0],loai[1]])

plt.xlabel("Nhãn báo")
plt.ylabel("Số lượng")
plt.title("Thống kê từng loại theo nhãn báo")
plt.show()
```

```python
plt.figure(figsize=(5.5,5.5))
data = [2118,2028]
graph = plt.bar(["Thật","giả"],data)
plt.title('Thống kê từng loại theo nhãn báo trong tập dữ liệu validate')
plt.xlabel("Nhãn báo")
plt.ylabel("Số lượng")

i = 0
for p in graph:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy()
    plt.text(x+width/2,
             y+height*1.01,
             str(data[i]),
             ha='center',
             weight='bold')
    i+=1
plt.show()
```
